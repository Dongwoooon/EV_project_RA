## 2. Extracting matched rows

### making class list
making list with the numbers of class column

```python
import pandas as pd
import zipfile
import csv

zf_patent1 = zipfile.ZipFile('E:/patent/all_patent/harvard_dataset_final_merged_output_1.zip') 
csv_patent1 = zf_patent1.open('harvard_dataset_final_merged_output_1.csv')
csv_patent_reader1 = csv.reader(csv_patent1)

header = csv_patent_reader1.next()
mainclass_col = []

for j in range(62457,63354):
    if 'mainclass' in header[j]:   
        mainclass_col.append(j)
```

making csv file with vehicle class

```python
classinfo = pd.read_csv('E:/patent/veh_class.csv')
veh_class = classinfo['class'].tolist()
veh_class = map(int,veh_class)
```

```python
myfile1 = open("harvard_vehicle_patent1.csv", 'wb')
myfile_writer1 = csv.writer(myfile1,quoting=csv.QUOTE_ALL)
myfile_writer1.writerow(header)

for line in csv_patent_reader1:
    for i in mainclass_col:
        if line[i]=='':
                break
                
        else:
            try:            
                if int(line[i]) in veh_class:
                    myfile_writer1.writerow(line)
                    break
                
            except ValueError:
                pass                
        
       
myfile1.close()
```

```python
zf_patent2 = zipfile.ZipFile('E:/patent/all_patent/harvard_dataset_final_merged_output_2.zip')
csv_patent2 = zf_patent2.open('harvard_dataset_final_merged_output_2.csv')
csv_patent_reader2 = csv.reader(csv_patent2)

csv_patent_reader2.next()
myfile2 = open("harvard_vehicle_patent2.csv", 'wb')
myfile_writer2 = csv.writer(myfile2,quoting=csv.QUOTE_ALL)
myfile_writer2.writerow(header)

for line in csv_patent_reader2:
    for i in mainclass_col:
        if line[i]=='':
                break
                
        else:
            try:            
                if int(line[i]) in veh_class:
                    myfile_writer2.writerow(line)
                    break
                
            except ValueError:
                pass                
        
       
myfile2.close()
```
